```python
# load necessary libraries 
import numpy as np
import pandas as pd
from datetime import datetime
```

```python
# Read The Metropolitan Museum of Art Open Access CSV
met_objects = pd.read_csv('MetObjects.csv')
met_objects.head()
```

```python
# Get rows and column details
met_objects.shape
```

```python
# Find out the data types 
met_objects.info()
```

```python
# Find missing value for all the columns of the dataframe
met_objects.isna().sum().sort_values(ascending = False)
```

```python
# We will now identify the duplicates
met_objects.duplicated(['Object Number']).sum()
```

```python
# Clean or Remove the duplicates and gets it rows and column details
Nodup_met_objects = met_objects.drop_duplicates(subset='Object Number')
Nodup_met_objects.shape
```

```python
# Hierarchical index
# Create a subset dataframe 
subset_objects = Nodup_met_objects[['Object Number', 'Department', 'Title', 'Object Name', 'Medium', 'Classification', 'Is Public Domain']]

# Select random sample of 2000 rows
random_objects = subset_objects.sample(2000)

# Reset Index of random dataframe
random_objects = random_objects.reset_index()
random_objects.head()
```

```python
# Create Hierarchical indexing using set_index
random_H_objects = random_objects.set_index(['Department', 'Medium'])
random_H_objects.head()
```

```python
# Reshaping the sample dataframe
stack_objects = random_objects.stack()
stack_objects
```

```python
# We will now reshape the rows into the columns using unstack
stack_objects.unstack()
```

```python
# Grouping 
# Check the count in either values for "Is public domain"

group_objects = random_objects.set_index(['Is Public Domain'])
group_objects.groupby(bool).count()
```

```python
# Crosstab
# Summarize the total records by calssifications and departments, and display top 10 classifications

crosstab_objects = pd.crosstab(subset_objects.Classification, subset_objects.Department, margins = True)
crosstab_objects.sort_values('All', ascending=False).head(11)
```

```python
# Find the type of a date field
type(Nodup_met_objects['Object Begin Date'])
```

```python
# Convert the Object Begin date to a datetime format and Generate a date range
datetime_objects = pd.to_datetime(Nodup_met_objects['Object Begin Date'])
daterange = pd.date_range(datetime.today(), periods=100).tolist()
daterange
```

```python
# Create timestamp
tmstp = pd.Timestamp(year = 2020,  month = 10, day = 26,  
                  hour = 20, minute = 30, second = 30, tz = 'US/Central')
tmstp
```

```python
# Convert timestamp to period
period = tmstp.to_period(freq ='M')
period
```

```python
# Convert period back to timestamp
period.to_timestamp(freq ='M') 
```
